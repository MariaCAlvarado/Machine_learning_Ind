```python
from tensorflow.keras.models import load_model
```

```python
BASE_FOLDER = '/content/drive/My Drive/Redes_neuronales_y_deep_learning/'
#pre_trained_model.save(BASE_FOLDER + "Cacao_cnn.f5")
modelo = load_model(BASE_FOLDER + "Cacao_cnn.f5")
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
def process_image(path):
    imagen = Image.open(path)
    imagen = imagen.resize((224,224))
    imagen = np.array(imagen)
    imagen = imagen/255.0
    return imagen


def prediccion(path):
  dicc_im = {0:'Fito',1:'Monilia',2:'Sana'}
  imagen = process_image(path)
  prediccion = modelo.predict(np.expand_dims(imagen, axis=0))
  plt.imshow(imagen)
  plt.axis("off")
  plt.title(dicc_im[np.argmax(prediccion)])
  plt.show()
  print(prediccion)
  return np.argmax(prediccion)


```

```python
path ='/content/monilia.jpg'
pred = prediccion(path)
```
